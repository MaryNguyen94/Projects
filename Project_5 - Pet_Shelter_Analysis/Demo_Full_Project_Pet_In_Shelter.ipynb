- Question 4: If spay/neuter for a dog costs 100 and a spay/neuter for a cat costs 50, how much did the shelter spend overtime on these procedures?

```python
# Step 1: Slice data frame
short_intake = intake[['animal_id', 'sex_upon_intake']]
short_outcome = outcome[['animal_id', 'sex_upon_outcome', 'datetime', 'animal_type']]
# Drop duplicate values in intake dataframe to see that animal's information
left_short_intake = short_intake.drop_duplicates(subset = ['animal_id'])
new_short_outcome = short_outcome.sort_values(by = 'datetime')
right_short_outcome = new_short_outcome.drop_duplicates(subset = ['animal_id'], keep = 'last')
right_short_outcome
#intersection 2 data frame to see columns values
## .pd.merge()
intersection_data = pd.merge(left_short_intake, right_short_outcome, how = 'inner', on = ['animal_id'])
intersection_data
# Step 2: find the animal that had not sp/ne before in sex_upon_intake column
intake_no_spne = intersection_data[(intersection_data.sex_upon_intake != 'Spayed Female') &
                                   (intersection_data.sex_upon_intake != 'Neutered Male')]
intake_no_spne
# Step 3: find the animal that sp/ne in sex_upon_outcome column
# Step 4: find the animal that sp/ne by staff
outcome_with_spne = intake_no_spne[(intake_no_spne.sex_upon_outcome == 'Spayed Female') |
                                   (intake_no_spne.sex_upon_outcome == 'Neutered Male')]

########################
# groupby()
Number_SpNe = outcome_with_spne.groupby(outcome_with_spne.animal_type).count()
Number_SpNe_Cat = Number_SpNe.loc['Cat'].animal_id
Number_SpNe_Dog = Number_SpNe.loc['Dog'].animal_id
Number_SpNe_Dog * 100 + Number_SpNe_Cat * 50
```

- Question 5: What are the adoption rates for the following age groups?
<br>baby: 4 months and less
<br>young: 5 months - 2 years
<br>adult: 3 years - 10 years
<br>senior: 11+

```python
# Step 1: Find the animal had adoption in outcome_type
adoption_outcome =  outcome[outcome.outcome_type == 'Adoption']

# Step 2: Drop NA values
adoption_outcome = adoption_outcome[adoption_outcome['age_upon_outcome'].notna()]

# Step 3: Find the data contain 'month' charater
age_upon_months = adoption_outcome[adoption_outcome['age_upon_outcome'].str.contains('month')]

# Step 4: Find the data contain 'year' charater
age_upon_years = adoption_outcome[adoption_outcome['age_upon_outcome'].str.contains("year")]

# Step 5: Combine 2 data frame above, we have animal data upper 1 month
Upper_1_month = pd.concat([age_upon_months, age_upon_years])

# Step 6: Subtract 2 datafames to get animal data below 1 month
Lower_1_month = pd.concat([adoption_outcome,Upper_1_month]).drop_duplicates(keep=False)

# Step 7: Filter the data between baby, young, aldult and senior animal by index number.

list_Baby_index = []
list_Young_index = []
list_Aldult_index = []
list_Senior_index = []

# Find data in age_upon_months datafame
# Step 7.1: Between 1 month and 1 year
for i in range(age_upon_months.shape[0]):
    List_separate_values = age_upon_months.age_upon_outcome.iloc[i].split(' ')
    if int(List_separate_values[0]) >= 5:
        list_Young_index.append(age_upon_months.age_upon_outcome.index[i])
    else:
        list_Baby_index.append(age_upon_months.age_upon_outcome.index[i])
        
# Step 7.2: Upper 1 year
for i in range(age_upon_years.shape[0]):
    List_separate_values = age_upon_years.age_upon_outcome.iloc[i].split(' ')
    if int(List_separate_values[0]) < 3:
        list_Young_index.append(age_upon_years.age_upon_outcome.index[i])
    elif 3 <= int(List_separate_values[0]) <= 10:
        list_Aldult_index.append(age_upon_years.age_upon_outcome.index[i])
    else: 
        list_Senior_index.append(age_upon_years.age_upon_outcome.index[i])
        
# Step 8: Show dataframe
Babyrate = round(100 / adoption_outcome.shape[0] * int(len(list_Baby_index) + Lower_1_month.shape[0]),3)
Youngrate = round(100 / adoption_outcome.shape[0] * int(len(list_Young_index)),3)
Aldultrate = round(100 / adoption_outcome.shape[0] * int(len(list_Aldult_index)),3)
Seniorrate = round(100 / adoption_outcome.shape[0] * int(len(list_Senior_index)),3)

Data = [['Baby rate', Babyrate], ['Young rate', Youngrate], ['Aldult rate', Aldultrate], ['Senior rate', Seniorrate]]
df = pd.DataFrame(Data, columns=['Age_Group', 'Adoption_rates'])
```
