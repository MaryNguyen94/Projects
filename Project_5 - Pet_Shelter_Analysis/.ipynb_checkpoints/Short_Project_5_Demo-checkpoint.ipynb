# Introduce to numpy

```python
import numpy as np

# Sample age data
ages = np.random.randint(100,size = 10)
# Convert the list to a NumPy array
age_array = np.array(ages)
mean_age = np.mean(age_array)
print("Mean Age:", mean_age)
median_age = np.median(age_array)
print("Median Age:", median_age)
std_dev = np.std(age_array)
print("Standard Deviation:", std_dev)
min_age = np.min(age_array)
max_age = np.max(age_array)
print("Minimum Age:", min_age)
print("Maximum Age:", max_age)
```

```python
# Imagine you have dataframe like this

# initialize list elements
data1 = [10,20,30,40,50,60,20,40,90,100]
data2 = ['alpha', 'beta', 'gramma', 'beta', 'gramma', 'omega', 'omega', 'delta','omega','omega']
# Create the pandas DataFrame with column name is provided explicitly
df = pd.DataFrame({'Number': data1, 'Category': data2})
# print dataframe.
df
```

```python
# Select all values in column 'Category'
dup_cate = df.Category
print(dup_cate)
print('_______________________________________1')

# Use '.duplicated()' to find the duplicate data
print(dup_cate.duplicated())
print('_______________________________________2')

# As you can see, number 20 and 40 is duplicate, so these row return 'True'
# Combine with '.isin' to find values that match previous query results
print(dup_cate.isin(dup_cate[dup_cate.duplicated()]))
print('_______________________________________3')

# And choose the dataframe we want to filter
dup_data = df[dup_cate.isin(dup_cate[dup_cate.duplicated()])]
print(dup_data)
print('_______________________________________4')

# Now, we can count total rows has duplicate
print('Threr are', dup_data.shape[1],'columns in dataset')
print('There are',dup_data.shape[0],'rows has duplicate')
print('_______________________________________5')

# Or check what category is duplicate
print('There are an array of duplicate category', dup_data.Category.unique())
print('_______________________________________6')

# And count how many time each category are duplicate by combine 'groupby' and 'count' method
# Try 'sum' Instead of 'count'
print(dup_data.groupby(by = 'Category').count())
print('_______________________________________7')

# We can use .sort_values to show descending values
print(dup_data.groupby(by = 'Category').count().sort_values(by = 'Category', ascending = False))
```
